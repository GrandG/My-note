## centos7上安装python3
1. ```sudo mkdir /usr/local/python3``` # 创建安装目录
2. ```wget --no-check-certificate https://www.python.org/ftp/python/3.6.0/Python-3.6.0.tgz``` # 下载python源
3. ```tar -zxvf Python-3.6.0.tgz``` # 解压缩
4. ```cd Python-3.6.0``
5. ```yum install gcc``` # 安装c编译器;  ```yum install make```
7. ```yum -y install zlib*``` 安装依赖包
8. ```yum install openssl-devel``` # 必须安装这个, 否则pip install 无法下载包
9. ```yum install sqlite-devel``` # 安装sqlite
10. ```sudo ./configure --prefix=/usr/local/python3``` # 指定创建目录
11. ```sudo make```
12. ```sudo make install```
13. ```sudo ln -s /usr/local/python3/bin/python3 /usr/bin/python3``` # 建立软连接, 这时候输入python3就可以打开python3了

## 安装pip
1. ```wget --no-check-certificate https://github.com/pypa/pip/archive/9.0.1.tar.gz```
2. ```tar -zvxf 9.0.1 -C pip-9.0.1```
3. ```cd pip-9.0.1```
4. ```python3 setup.py install```
5. ```udo ln -s /usr/local/python3/bin/pip /usr/bin/pip3``` # 建立软连接
6. ```pip3 install --upgrade pip```  # 更新

## django调试服务器开启后, 访问地址访问不了
- 原因是防火墙没有开启相应的端口
- 开启相应的端口, 再重启防护墙就可以了
- 同时再setting.py的ALLOW_HOST中要加上服务器所在主机的地址, 也可以写成'*', 即全部地址允许
- 相关命令
    * ```firewall-cmd --list-ports```, 查看所有开放的端口, 如果没有开放的端口, 无任何显示
    * ```firewall-cmd --query-port=80/tcp```, 产看80端口是否开启
    * ```firewall-cmd --permanent --zone=public --add-port=8000/tcp```, 开放8000端口
    * ```firewall-cmd --permanent --zone=public --remove-port=8000/tcp```, 移除8000端口
    * ```systemctl start/stop/restart firewalld.service```, 开启,关闭, 重启防火墙
- 有个小坑是, 我的项目设置git屏蔽了migrations文件夹,所以拉取下来后没有这个文件夹, 导致迁移失败
