## 虚拟机硬件管理

#### 网络适配器

安装完虚拟机后, 可以在网络连接里看到多了两块网卡: VMnet1, VMnet8

- 桥接: 虚拟机和真实机直接的通信, 使用真实的本地网卡, 缺点是虚拟机会占用真实网段的ip. 使用这种方法虚拟机可以和局域网中的其他设备通信. 推荐用这个
- NAT: 使用VMnet.8这块网卡与真实机通信, 只能与真实机通信, 不能与局域网中其他设备通信; 可以上网
- Host-only: 使用VMnet.1这块网卡与真实机通信, 只能与真实机通信, 不能与局域网中其他设备通信; 不可以上网

## Linux系统分区

#### 分区类型

- 在Linux下与Windows下规则一样, 规则由硬盘结构决定.
- 主分区: 最多只能4个
    + 这是由硬盘结构决定的. 硬盘物理上分成4个扇区, 每个扇区有512Bytes; 512Bytes当中446Bytes用来记录启动信息, 64Bytes用来进行分区表示, 剩下2Bytes一定是55 AA(十六进制).64Bytes当中, 每16Bytes表示一个分区, 所以最多只能表示4个分区.
- 扩展分区:
    + 最多只能有1个
    + 主分区加扩展分区最多只能有4个
    + 不能写入数据, 只能包含逻辑分区
- 逻辑分区
    + 逻辑分区能分多少个呢? 在windows中用a-z作为盘符命名, 其中a, b作为软驱的盘符, c作为主分区1的盘符,所以理论上能再分23个逻辑分区
    
#### 格式化
- 格式化又称逻辑格式化, 他是指用户选定的文件系统(如FAT16, FAT32, NTFS, EXT2, EXT3, EXT4等)
- 分好区后, 还要格式化才能读写信息
- 清楚数据不是格式化的目的, 只是格式化带来的后果
- 格式化主要干了两件事:
    + 格式化把分区划分成等大小的block
    + 在分区列表中建立一个二维表格, 记录文件的id, 权限, 修改时间, 存储的block的位置. 其中id号叫做**I节点**(I node)
    
#### Windows与Linux区别

- Windows只要:
    + 分区, 
    + 格式化, 
    + 分配盘符名就可以; 
- Linux需要
    + 分区, 
    + 格式化, 
    + 给每个分区起一个**设备文件名**, 
    + 分配盘符
        * 其实盘符这个概念是Windows的, Linux下应该叫**挂载点**

#### 硬件设备文件名

- 在Linux中一切皆文件, 其中硬件的文件的文件名, 叫做硬件文件名
- **/**指的是根目录
- **/dev**保存所有的硬件设备文件

硬件         |设备文件名
-----------------|----------
IDE硬盘       |/dev/hd[a-d]
SCSI/SATA/USB硬盘|/dev/sd[a-p]
光驱         |/dev/cdrom或/dev/hdc
软盘         |/dev/fd[0-1]
打印机(25针)    |/dev/lp[0-2]
打印机(USB)     |/dev/usb/lp[0-15]
鼠标         |/dev/mouse

#### 分区设备文件名

- 设备文件名
    + /dev/hda1    (IDE硬盘接口), 这里是说第一块(a的意思)IDE接口硬盘的的第一个(1的意思)分区
    + /dev/sda1    (SCSI硬盘接口, SATA硬盘接口)
    + 注意, 逻辑分区的分区号一定是从5开始, 即使只分了不够3个的主分区也是; 1, 2, 3, 4只能给主分区或扩展分区使用
    + [这里](https://www.bilibili.com/video/av4913985/#page=10)介绍了不同接口硬盘的区别
    
#### 挂载

- Linux下的**挂载点**等同于WIndows下的**盘符**, 分配挂载点的过程称为**挂载**
- 但是Linux下不是用a-z作为盘符, 而是用目录作为盘符
    + 必须分区
        * / (根分区)
        * swap分区(交换分区, 内存是4GB以下的, 分为内存的两倍, 内存是4GB以上的, 与内存一样大), 即虚拟内存
    + 推荐分区
        * /boot (启动分区, 200MB), 存放系统的启动信息, 如果不分, /boot将使用根目录空间, 一旦根目录写满后, 系统将不能启动
        
#### 文件系统结构

- **从树形结构上来看, /boot, /etc, /home这些都在/下面, 但从硬盘空间上来看, 可以把这些目录分配给不同的硬盘空间**

## Linux系统安装

- Linux不通过主机名在局域网中寻找位置, WIndows才会, 所以Linux主机名可以重名
- swap不是目录, 不存在/swap的目录, swap是文件系统类型, 与fat32, ext4这些是并列的, swap分区没有挂载点, 是系统内部调用的
- Linux安装方式
    + Desktop (桌面)
    + Minimal Desktop (最小化桌面)
    + Minimal (最小化, 在生产服务器中使用这个)
    + Basic Server (基本服务器, 在本次实验中用这个)
    + Database Server (数据库服务器)
    + Web Server (网页服务器)
    + Virtual Host (虚拟主机)
    + software development workstation (软件开发工作站)
 - 安装日志
     + /root/install.log: 存储了安装在系统中的软件包机器版本信息
     + /root/install.log.syslog: 存储了在安装过程中留下的事件记录
     + /root/anaconda-ks.csf: 以Kickstart配置文件的格式记录安装过程中设置的选项信息
